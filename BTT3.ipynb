```python
a,b,m=map(float,input().split())
c,d,n=map(float,input().split())
d=a*d-c*b
dx=n*d-c*m
dy=a*n-m*b
if d!=0:
    x=dx/d
    y=dy/d
    print("He phuong trinh co nghiem duy nhat (x,y)=(",x,",",y,")")
elif dx==0 and dy==0:
    print("He phuong trinh co vo so nghiem")
else:
    print("He phuong trinh vo nghiem")
```
