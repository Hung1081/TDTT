```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

a, b = map(int, input().split())
prime_sum = sum(x for x in range(a, b + 1) if is_prime(x))
print(prime_sum)
```

```python
def check(n):
    ans=1
    if n==1:
        return ans
    while n>1:
        if n%2==0:
            n=n/2
            ans+=1
        else:
            n=3*n+1
            ans+=1
    return ans

n=int(input("Nhap so nguyen duong: "))
L=1
ans=1
for x in range(1,n+1):
    if L<check(x):
        L=check(x)
        ans=x
print(ans,L)
        
```

```python
def tonguoc(n):
    s=0
    i=1
    while i*i<=n:
        if n%i==0:
            s+=i
            if i!=n//i:
                s+=n//i
        i+=1
    return s-n
a,b=map(int,input().split())
if tonguoc(a)==b and tonguoc(b)==a:
    print(True)
else:
    print(False)    
```

```python
def ucln(a,b):
    x=min(a,b)
    if b%a==0 or a%b==0:
        return a
    else:
        i=1
        ans=0
        while i*i<=x:
            if x%i==0:
                ans=max(i,ans)
                k=x//i
                if a%k==0 and b%k==0:
                    ans=max(ans,k)
            i+=1
        return ans
a,b=map(int,input().split())
print(ucln(a,b))
```
