```python
s=list(map(int,input().split()))
k=int(input())
ans=[]
for x in s:
    if (k-x) in s:
        if (x,k-x) not in ans:
            ans.append((x, k-x))
print(ans)
```

```python
dic={}
s=list(map(str,input().split()))
n=int(input())
for i in range(0,len(s),2):
    dic[s[i]]=int(s[i+1])
ans={}
for x in dic:
    if dic[x]>=n:
        ans[x]=dic[x]
print(ans)
```
