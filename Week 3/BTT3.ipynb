W3A16

```python
x = float(input())

if x >= 0:
    down = int(x)
else:
    down = int(x) if x == int(x) else int(x) - 1

if x <= 0:
    up = int(x)
else:
    up = int(x) if x == int(x) else int(x) + 1
if x - int(x) < 0.5 and x - int(x) > -0.5:
    near = int(x)
elif x >= 0:
    near = int(x) + 1
else:
    near = int(x) - 1

print(down, up, near)
```
