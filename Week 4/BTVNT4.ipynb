```python
def uoc(n):
    s=[]
    i=1
    while i*i<=n:
        if n%i==0:
            s.append(i)
            if i*i!=n:
                s.append(n//i)
        i+=1
    s=sorted(s)
    return s    
n=int(input("Nhập số nguyên: "))
print(uoc(n))
```

```python
def demuoc(n):
    s=1
    for i in range(2,int(n**0.5)+1):
        t=1
        while n%i==0:
            n//=i
            t+=1
        s*=t
    if n>1:
        s*=2
    return s
n=int(input("Nhập số nguyên: "))
print(demuoc(n))
```

```python
def uc(a,b):
    i=2
    ans=[1]
    x=min(a,b)
    if a%b==0 or b%a==0:
        ans.append(x)
    while i*i<=x:
        if a%i==0 and b%i==0:
            ans.append(i)
            if a%(x//i)==0 and b%(x//i)==0:
                ans.append(x//i)
        i+=1
    ans=sorted(ans)
    return ans
a,b=map(int,input().split())
print(uc(a,b))
```

```python
def tonguoc(n):
    s=0
    i=1
    while i*i<=n:
        if n%i==0:
            s+=i
            if i!=n//i:
                s+=n//i
        i+=1
    return s-n
n=int(input("Nhập số nguyên: "))
if tonguoc(n)==n:
    print("YES")
else:
    print("NO")
    
```
