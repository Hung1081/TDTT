```python
a = tuple(map(float, input().split()))
b = tuple(map(float, input().split()))
c = tuple(map(float, input().split()))
d = tuple(map(float, input().split()))
p = tuple(map(float, input().split()))

def area(a, b, c):
    return abs((b[0] - a[0]) * (c[1] - a[1]) - (c[0] - a[0]) * (b[1] - a[1])) / 2

def check(a, b, c, d, p):
    tongdt = area(a, b, c) + area(a, c, d)
    if tongdt <= (area(a, b, p) + area(b, c, p) + area(c, d, p) + area(d, a, p)):
        return True
    else:
        return False

print(check(a, b, c, d, p))
```

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    """
    Tính tiền taxi: mở cửa, km, chờ, đêm, giảm giá.
    """
    fare = 0
    
    # 1. Thiết lập bảng giá
    if car_type == "4_seat":
        open_price = 10000
        km1 = 15000 # 1-10km
        km2 = 12000 # >10km
    elif car_type == "7_seat":
        open_price = 12000
        km1 = 17000
        km2 = 14000
    else:
        return "Loại xe không hợp lệ"
        
    # 2. Tính tiền theo km
    if km <= 1:
        fare = open_price
    elif km <= 10:
        fare = open_price + (km - 1) * km1
    else:
        fare = open_price + 9 * km1 + (km - 10) * km2
        
    # 3. Phí chờ (500đ/phút)
    fare += wait_minutes * 500
    
    # 4. Phụ phí ban đêm (22h - 6h)
    if start_hour >= 22 or start_hour < 6:
        fare *= 1.1
        
    # 5. Giảm giá (nếu > 100km)
    if km > 100:
        fare *= 0.9
        
    return round(fare, -3) # Làm tròn nghìn đồng

# --- Ví dụ sử dụng ---
# Đi 120km, xe 7 chỗ, 11h đêm (23h), chờ 10 phút
total = smart_taxi_fare("7_seat", 23, 120, 10)
print(f"Tổng tiền: {total:,} VND")
```
