```python
def bs(a,x):
    h=len(a)-1
    l=0
    while l<=h:
        m=(l+h)//2
        if a[m]==x:
            return m
        elif a[m]>x:
            h=m-1
        else:
            l=m+1
    return -1

a=list(map(int,input().split()))
a.sort()
x=int(input())
print(bs(a,x))

```

```python
def bbs(a):
    for i in range(len(a)-1):
        for j in range(len(a)-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a
def ss(a):
    for i in range(len(a)-1):
        min_idx=i
        for j in range(i+1,len(a)):
            if a[j]<a[min_idx]:
                min_idx=j
        a[i],a[min_idx]=a[min_idx],a[i]
    return a
a=list(map(int,input().split()))
b=a.copy()
print(bbs(a))
print(ss(b))
```

```python
def giai_nhanh(A, B, C, D):
    danh_sach = [('A', A), ('B', B), ('C', C), ('D', D)]
    
    # Quy tắc phân loại thông minh:
    # - Nếu là số (int hoặc float): Giữ nguyên số để xếp theo giá trị toán học.
    # - Nếu là chữ (str): Chuyển về chữ thường để xếp theo bảng chữ cái.
    def quy_tac(x):
        gia_tri = x[1]
        if isinstance(gia_tri, (int, float)):
            return (gia_tri, "") # Số đứng trước, không cần phân biệt hoa thường
        else:
            return (str(gia_tri).lower(), str(gia_tri)) # Chữ xếp theo từ điển và hoa/thường

    # Sắp xếp mảng dựa trên quy tắc thông minh trên
    danh_sach.sort(key=quy_tac)
    
    # Trả về kết quả, các chữ cái cách nhau bằng dấu cách
    return " ".join([x[0] for x in danh_sach])


# === KIỂM TRA VỚI SỐ NHIỀU CHỮ SỐ (Đã chính xác) ===
print(giai_nhanh(A=12, B=9, C=120, D=5))
# Máy tính hiểu đúng: 5 -> 9 -> 12 -> 120
# Kết quả in ra: D B A C

# === KIỂM TRA VỚI CHỮ (Vẫn chính xác) ===
print(giai_nhanh(A="rau muống", B="Cà rốt", C="Rau dền", D="rau dền"))
# Kết quả in ra: B C D A
```
