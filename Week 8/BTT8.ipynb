```python
class ht:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def area(self):
        return 3.14*self.r*self.r
    def volume(self):
        return 3.14*self.r*self.r*self.h
r,h=map(float,input().split())
a=ht(r,h)
print(a.area(),a.volume())
```

```python
class date:
    def __init__(self,day,month,year):
        self.day=day
        self.month=month
        self.year=year
    def display(self):
        if self.day>31 or self.month>12:
            return "Invalid date"
        match self.month:
            case 1|3|5|7|8|10:
                if self.day==31:
                    return f"{1}/{self.month+1}/{self.year}"
            case 2:
                if self.day==28:
                    return f"{1}/{self.month+1}/{self.year}"
            case 4|6|9|11:
                if self.day==30:
                    return f"{1}/{self.month+1}/{self.year}"
            case 12:
                if self.day==31:
                    return f"{1}/{1}/{self.year+1}"
        return f"{self.day+1}/{self.month}/{self.year}"
a,b,c=map(int,input().split())
d=date(a,b,c)
print(d.display())
```

```python
class vt:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def check(self):
        if self.x==0 and self.y==0:
            return "Origin"
        elif self.x==0:
            return "Y-axis"
        elif self.y==0:
            return "X-axis"
        else:
            return (self.x**2+self.y**2)**0.5
a,b=map(float,input().split())
v=vt(a,b)
print(v.check())
        
```

```python
class Calculator:
    def __init__(self, num1, num2):
        """Khởi tạo lớp với 2 thuộc tính số ban đầu"""
        self.num1 = num1
        self.num2 = num2

    def set_numbers(self, num1, num2):
        """Phương thức 7: Cập nhật 2 số mới"""
        self.num1 = num1
        self.num2 = num2
        print(f"-> Đã cập nhật 2 số mới: {self.num1} và {self.num2}")

    def add(self):
        """Phương thức 1: Cộng"""
        return self.num1 + self.num2

    def subtract(self):
        """Phương thức 2: Trừ"""
        return self.num1 - self.num2

    def product(self):
        """Phương thức 3: Nhân"""
        return self.num1 * self.num2

    def divide(self):
        """Phương thức 4: Chia"""
        if self.num2 == 0:
            return "Lỗi: Không thể chia cho 0!"
        return self.num1 / self.num2

    def power(self):
        """Phương thức 5: Hàm mũ (num1 lũy thừa num2)"""
        return self.num1 ** self.num2

    def mod(self):
        """Phương thức 6: Lấy phần dư"""
        if self.num2 == 0:
            return "Lỗi: Không thể chia lấy dư cho 0!"
        return self.num1 % self.num2


# --- CHƯƠNG TRÌNH CHÍNH (MENU ĐIỀU KHIỂN) ---
print("=== CHƯƠNG TRÌNH MÁY TÍNH BỎ TÚI ===")
# Nhập 2 số ban đầu để khởi tạo object
n1 = float(input("Nhập số thứ nhất: "))
n2 = float(input("Nhập số thứ hai: "))
calc = Calculator(n1, n2)

while True:
    print("\n" + "="*30)
    print(f"Hiện tại: Số thứ nhất = {calc.num1} | Số thứ hai = {calc.num2}")
    print("CHỌN PHÉP TOÁN:")
    print("1. Cộng (add)")
    print("2. Trừ (subtract)")
    print("3. Nhân (product)")
    print("4. Chia (divide)")
    print("5. Hàm mũ (power)")
    print("6. Lấy phần dư (mod)")
    print("7. Đổi 2 số mới (set_numbers)")
    print("="*30)
    
    choice = input("Nhập lựa chọn của bạn (1-7): ")

    # Xử lý các lựa chọn phép tính
    match choice:
        case "1":
            print(f"Kết quả {calc.num1} + {calc.num2} = {calc.add()}")
        case "2":
            print(f"Kết quả {calc.num1} - {calc.num2} = {calc.subtract()}")
        case "3":
            print(f"Kết quả {calc.num1} * {calc.num2} = {calc.product()}")
        case "4":
            print(f"Kết quả {calc.num1} / {calc.num2} = {calc.divide()}")
        case "5":
            print(f"Kết quả {calc.num1} ^ {calc.num2} = {calc.power()}")
        case "6":
            print(f"Kết quả {calc.num1} % {calc.num2} = {calc.mod()}")
        case "7":
            new_n1 = float(input("Nhập số thứ nhất mới: "))
            new_n2 = float(input("Nhập số thứ hai mới: "))
            calc.set_numbers(new_n1, new_n2)
        case _:
            print("Lựa chọn không hợp lệ! Vui lòng chọn từ 1 đến 7.")
            continue # Quay lại đầu vòng lặp, bỏ qua câu hỏi thoát

    # Hỏi người dùng có muốn thoát không
    exit_prompt = input("\nBạn có muốn thoát? (Gõ 'yes'/'Yes'/'YES' để thoát): ")
    if exit_prompt in ["yes", "Yes", "YES"]:
        print("Tạm biệt!")
        break
```

```python
class cc:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f"({self.x},{self.y})"
    def __add__(self,other):
        return cc(self.x+other.x,self.y+other.y)
    def __sub__(self,other):
        return cc(self.x-other.x,self.y-other.y)
    def __mul__(self,other):
        return cc(self.x*other.x,self.y*other.y)
    def __truediv__(self,other):
        return cc(self.x/other.x,self.y/other.y)
    def __floordiv__(self,other):
        return cc(self.x//other.x,self.y//other.y)
    def __mod__(self,other):
        return cc(self.x%other.x,self.y%other.y)
    def __pow__(self,other):
        return cc(self.x**other.x,self.y**other.y)
    def update(self,x,y):
        self.x=x
        self.y=y
while True:
    a,b=map(str,input("Ban co muon thoat khong :").split())
    if a=="yes" or a=="Yes" or a=="YES":
        print("Tam biet")
        break
    else:
        a,b=float(a),float(b)
        c,d=map(float,input("Nhap hai so moi: ").split())
        m=cc(a,b)
        n=cc(c,d)
        print(m+n,m-n,m*n,m/n,m//n,m%n,m**n)
```

```python
class shopping:
    def __init__(self):
        self.box=[]
    def __str__(self):
        return str(self.box)
    def addit(self,name,price):
        self.box.append((name,price))
    def removeit(self,name):
        for i in self.box:
            if i[0]==name:
                self.box.remove(i)
                return
        print("Khong tim thay san pham")
    def isempty(self):
        return len(self.box)==0
    def sumit(self):
        return sum(i[1] for i in self.box)
    def total(self):
        return self.box
    def removeall(self):
        self.box.clear()
s=shopping()
s.addit("Ao",100)
s.addit("Quan",200)
s.addit("Giay",300)
print(s.removeit("Ao"))
print(s.isempty())
print(s.sumit())
print(s.total())
s.removeall()
print(s)
```
